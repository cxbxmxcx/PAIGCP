A Basic MNIST example

```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

import tensorflow.keras.models as models
import tensorflow.keras.layers as layers
```

Loading and normalizing data

```
(x_train, y_train) = np.asarray([[0,0],[1,0],[0,1],[1,1]]), np.asarray([[0],[1],[1],[0]])
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_train[:,0], y_train, x_train[:,1], c='r', marker='o')
```

```
print(x_train.shape)
print(y_train.shape)
```

Build the Model

```
model = models.Sequential()
model.add(layers.Dense(8, input_dim=2))
model.add(layers.Activation('tanh'))
model.add(layers.Dense(1))
model.add(layers.Activation('sigmoid'))

sgd = tf.keras.optimizers.SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd)

model.summary()
```

```
history = model.fit(x_train, y_train, batch_size=1, epochs=100)
```

```
model.predict(x_train[:1])
```

Train and evaluate

```
plt.plot(history.history['loss'], label='train loss') 
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='lower right')
```
